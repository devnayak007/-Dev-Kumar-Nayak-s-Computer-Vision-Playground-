```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
import tensorflow_hub as hub
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_img= load_image(r'/content/IMG_20240326_124814.jpg')
style_image = load_image(r'/content/monet.jpeg')
```

```python
plt.imshow(np.squeeze(content_img))
plt.show()
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylised_image = model(tf.constant(content_img),tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylised_image))
plt.show()
```

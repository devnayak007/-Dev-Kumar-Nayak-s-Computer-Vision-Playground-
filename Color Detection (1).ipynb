```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread(r"C:\Users\devna\Documents\Computer Vision\Image Processing\shapes.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
hsv_rgb = cv2.cvtColor(hsv,cv2.COLOR_BGR2RGB)
display= np.hstack((image_rgb,hsv_rgb))
plt.imshow(display)
```

```python
#Blue Color Triangle
lower_hue = np.array([65,0,0])
upper_hue = np.array([110,255,255])
mask =  cv2.inRange(hsv,lower_hue,upper_hue)
mask_rgb = cv2.cvtColor(mask,cv2.COLOR_BGR2RGB)
result = cv2.bitwise_and(image,image,mask=mask)
result_rgb = cv2.cvtColor(result,cv2.COLOR_BGR2RGB)
plt.imshow(result_rgb)
```

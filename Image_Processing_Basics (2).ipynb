#  Importing Dependancies

```python
import cv2
print(cv2.__version__)
from matplotlib import pyplot as plt
```

# Reading the Images

```python
image = cv2.imread(r'C:\Users\devna\Documents\Computer Vision\Image Processing\Batman.jpg')
image_rgb= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
print(image_rgb.shape)
```

# Converting into Gray Scale

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_rgb = cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
plt.imshow(gray_rgb)
```

```python
print(gray.shape)
```

# Converting into HSV File


```python
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
hsv_rgb = cv2.cvtColor(hsv,cv2.COLOR_BGR2RGB)
plt.imshow(hsv_rgb)
```

# Edge Detection

```python
canny_image = cv2.Canny(gray,60,60)
canny_image_rgb = cv2.cvtColor(canny_image,cv2.COLOR_BGR2RGB)
plt.imshow(canny_image_rgb)
```

# Erosion and Dilation

```python
#Dilation
import numpy as np
```

```python
kernel = np.ones((5,5),np.uint8)
dilate_image = cv2.dilate(canny_image,kernel,iterations=1)
dilate_image_rgb = cv2.cvtColor(dilate_image,cv2.COLOR_BGR2RGB)
plt.imshow(dilate_image_rgb)
```

```python
#erosion
eroded_image = cv2.erode(dilate_image,kernel,iterations=1)
eroded_image_rgb = cv2.cvtColor(eroded_image,cv2.COLOR_BGR2RGB)
plt.imshow(eroded_image_rgb)
```

```python
display = np.hstack((canny_image_rgb,dilate_image_rgb,eroded_image_rgb))
plt.imshow(display)
```

# Image De-Noising

```python
lion_image = cv2.imread(r'C:\Users\devna\Documents\Computer Vision\Image Processing\lion.jpg')
lion_image_rgb = cv2.cvtColor(lion_image,cv2.COLOR_BGR2RGB)
plt.imshow(lion_image_rgb)
```

```python
dst = cv2.fastNlMeansDenoisingColored(lion_image,None,20,20,10,15)
dst_rgb = cv2.cvtColor(dst,cv2.COLOR_BGR2RGB)
display_n= np.hstack((lion_image_rgb,dst_rgb))
plt.imshow(display_n)
```
